```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ruta_csv = '..\\data\\informacion_pacientes.csv'

df = pd.read_csv(ruta_csv,encoding='utf-8')
```

```python
df.head()
```

```python
df.shape
```

```python

conteo = df['numero_archivos'].value_counts()

ax = conteo.plot(kind='bar', color='skyblue', edgecolor='black')

for i, valor in enumerate(conteo):
    ax.text(i, valor + 0.5, str(int(valor)), ha='center', va='bottom')

plt.title('Cantidad de archivos por paciente')
plt.xlabel('Frecuencia')
plt.ylabel('Número de archivos')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
