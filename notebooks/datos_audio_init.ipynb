```python
import os
import pandas as pd
import wave

#Ruta donde se encuentran los archivos de audio
ruta_archivos = 'C:\\Users\\edaleon\\Downloads\\parinson_audio\\parinson_audio\\wavFiles'

#Los audios de los pacientes estn separados por carpeta
ids =  [nombre for nombre in os.listdir(ruta_archivos) if os.path.isdir(os.path.join(ruta_archivos, nombre))]

audios_data = []

#Se recorre cada subcarpeta para recoger la informacin de los audios grabados
for id in ids:
    audio_data = {}
    audio_id = os.path.join(ruta_archivos, id)
    audio_data['id'] = id
    # Verificar que la carpeta existe
    if os.path.isdir(audio_id):
        archivos_audio = [
            f for f in os.listdir(audio_id)
            if os.path.isfile(os.path.join(audio_id, f)) and
            os.path.splitext(f)[1].lower() in ['.wav','.WAV']]
                
        # Mostrar archivos
        for archivo in archivos_audio:

            try:
                with wave.open(os.path.join(audio_id,archivo), 'rb') as wav_file:
                    frames = wav_file.getnframes()
                    rate = wav_file.getframerate()
                    duracion = frames / float(rate)
            except Exception as e:
                duracion = None
                print(f"Error leyendo '{archivo}': {e}")
            if 'moca_p' in archivo.lower():
                audio_data['file_MoCA_P_Letter'] = archivo
                audio_data['lenght_MoCA_P_Letter'] = duracion
            elif 'animales' in archivo.lower():
                audio_data['file_Animales_Letter'] = archivo
                audio_data['lenght_Animales_Letter'] = duracion
            elif 'frutas' in archivo.lower():
                audio_data['file_Frutas_Letter'] = archivo
                audio_data['lenght_Frutas_Letter'] = duracion
            elif '_a_' in archivo.lower():
                audio_data['file_A_Letter'] = archivo
                audio_data['lenght_A_Letter'] = duracion
            elif '_f_' in archivo.lower():
                audio_data['file_F_Letter'] = archivo
                audio_data['lenght_F_Letter'] = duracion
            elif '_m_' in archivo.lower():
                audio_data['file_M_Letter'] = archivo
                audio_data['lenght_M_Letter'] = duracion
        audios_data.append(audio_data)
    else:
        print(f"La carpeta '{audio_id}' no existe.")

df_audios = pd.DataFrame(audios_data)
df_audios
```
